7. Derivative & Options Strategies
Use of options/futures can leverage views without margin, but carry risk of total loss.

7.1 Equity Options

Bullish Calls: Buy call options on strong stocks/indices (e.g. NASDAQ calls, or on individual tech stocks) to magnify upside. Example: a $5k purchase of call options expiring in 6–12 months could yield >10× if a stock doubles.

Bull Call Spreads: For cheaper exposure (buy-call/sell-higher-strike), capping max profit but limiting cost.

Covered Calls: Write covered calls on existing stock holdings to generate premium (income) if moderate growth expected.

Protective Puts: Hedge crash risk by buying put options on major indices (insurance).

Strategy: Use <10% of portfolio for speculative calls. For instance, in early 2025, buy calls on AI leader stocks anticipating tech rally. Sell options if volatility surges.

Risks: Options expire worthless if wrong. Only allocate what you can lose fully.

7.2 Commodity Futures/Options

Focus: Futures on oil, metals, or FX to speculate on big moves.

Why: High leverage (though margin required, but can be seen as high-risk alternative).

Strategy: Only for advanced traders. Example: going long on LNG futures if winter is harsh; shorting oil if oversupply news. Confine size (<5%).

Risks: Margin calls if markets turn. Not recommended for beginners.

7.3 Structured Products & Leverage ETFs (Limited)

Note: Leverage ETFs (2× Nasdaq, etc.) amplify returns but also losses. Since user requested no margin trading, these could be considered an alternative. E.g., SSO (2× SPY), UPRO (3× SPY). Use sparingly (a fraction of equity).

Strategy: If extremely bullish on a short-term move, a small use of 2× ETF can capture double returns. Always exit quickly on rollovers.

Risks: Decay due to daily rebalancing, huge losses if market corrects. Limit use.
